<h3>So what are Support Vector Machines? (SVMs)</h3> 

I thought it would be helpful to write some notes on the basics of support vector machines. 

The basic concept of SVM focuses on hyperplane separators in an n-dimensional space. What does this mean? Let's say we have a 3D chart of "o's" and "+'s" like the below graph. How would we go about separating the blue and red data points? 

The answer is using a hyperplane!

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def randrange(n, vmin, vmax):
    return (vmax-vmin)*np.random.rand(n) + vmin

fig = plt.figure(figsize=(8,6), edgecolor='k')
ax = fig.add_subplot(111, projection='3d')
n = 100
for c, m, zl, zh in [('r', 'o', -50, -25), ('b', '+', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zl, zh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')

plt.show()
```

```python
plt.plot([0,2,0], [0,0,2], 'ro')
plt.plot([4,2,4], [0,2,4], 'b+')
plt.axis([-1, 5, -1, 5])
plt.xlabel('x1')
plt.ylabel('x2')
```

Any easier way of looking at this is with a 2D hard margin example. 

Because this data is linearly separable, we can use a hard margin support vector machine to perform classification. This algorithm chooses one of the many different hyperplanes that separate the two classes. 

Intuitively, which hyperplane should we choose? 

####The best choice is the hyperplane that will leave the maximum margin for both classes.####

In this case, the hyperplane with the maximum margin would be equidistant between the red dot and the blue + that are closest to each other. The slope is -1.  To find the hyperplane we can use the support vectors which are the (2,0) (0,2) and (4,0) and (2,2). 

Because the weight vectors are normal to the hyperplane and we already established that the hyperplane has a slope of -1, we can set $\vec\omega = (c,c)$. Using the support vectors, we can set up the below system of equations.

__Equation using the first training point(blue +):__

$\vec\omega \ast (2,2) + b = 1$

$ 2c + 2c + b = 1$

$4c+b=1$

__Equation using the second training point(red dot):__

$\vec\omega \ast (2,0) + b = - 1$

$2c+b=-1$

When we subtract $4c+b=1$ from $2c+b=-1$ we get:

$2c =2$ where $c=1$

When we plug in $c$ we get $b =-3$ for both equations. 

###Solution $\vec\omega = (1,1)$ and $b = -3.$###

<h2>About the MNist Handwritten Data</h2>
<hr size=3>
The MNIST Dataset is a set of handwritten digit images. There are 60,000 examples for the training set and 10,000 examples for the testing set. I have randomly taken a few thousand examples for training and  testing. The images are stored as a matrix of a 28 X 28 pixels. Each pixel is a value between 0 and 255 where 0 is black and 255 is white and everything in between are shades of grey. You can find the original data set from <a href="http://yann.lecun.com/exdb/mnist/" target="_blank">Yann LeCun's </a> page.

##Let's take a look at the data:##

```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn import cross_validation
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
```

```python
test_data = pd.read_csv('http://irisabramson.com/dir/mnist_test.txt', header=None)
train_data= pd.read_csv('http://irisabramson.com/dir/mnist_train.txt', header=None)
```

## Linear Kernel

Step 1. Normalization using -1 to 1 where I set 0 to -1 and 255 to 1

```python
def norm(x):
    return (2*x/255)-1

scaled_data = train_data.ix[:,1:].apply(norm)
scaled_data_test = test_data.ix[:,1:].apply(norm)
```

Step 2. Split training data into validation folds

```python
kf = cross_validation.KFold(2000, n_folds=5, shuffle=True)
validation_params =  [.0001,.001,.01,.1,1,10,100,1000]

```

Step 3. SVM with Linear Kernel

```python

validation_results = np.zeros(len(validation_params))
for train_index, test_index in kf:
    X = scaled_data.ix[train_index,:]
    test_X = scaled_data.ix[test_index,:]
    y = train_data.ix[train_index,0]
    test_y =  train_data.ix[test_index,0]
    for cost in validation_params:
        clf = svm.SVC(C=cost, kernel="linear").fit(X,y)
        predictions = clf.predict(test_X)
        validation_results[validation_params.index(cost)] += accuracy_score(test_y, predictions)
        
```

Results of the Linear Kernel with resepct to each validation parameter

```python
.2*validation_results
```

## Gaussian Kernel

```python
kf = cross_validation.KFold(2000, n_folds=5, shuffle=True)
validation_params =  [10,100,1000,10000]

validation_results = np.zeros(len(validation_params))
for train_index, test_index in kf:
    X = scaled_data.ix[train_index,:]
    test_X = scaled_data.ix[test_index,:]
    y = train_data.ix[train_index,0]
    test_y =  train_data.ix[test_index,0]
    for cost in validation_params:
        clf = svm.SVC(C=cost, kernel="rbf").fit(X,y)
        predictions = clf.predict(test_X)
        validation_results[validation_params.index(cost)] += accuracy_score(test_y, predictions)
  
```

```python
.2*validation_results
```
